```python
## Task 1: Mark all launch sites on a map
# Setup cells - run these first
import folium
import pandas as pd

# Download and read the dataset
from js import fetch
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df = pd.read_csv(spacex_csv_file)

# Select relevant columns
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]

# Import necessary Folium components
from folium.plugins import MarkerCluster, MousePosition
from folium.features import DivIcon

# Define NASA coordinate
nasa_coordinate = [29.559684888503615, -95.0830971930759]

# Initial the map
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# For each launch site, add a Circle object based on its coordinate (Lat, Long) values
# In addition, add Launch site name as a popup label
for index, site in launch_sites_df.iterrows():
    # Get the coordinates for each launch site
    site_coordinate = [site['Lat'], site['Long']]
    
    # Add a circle for the launch site
    circle = folium.Circle(
        site_coordinate, 
        radius=1000, 
        color='#000000', 
        fill=True
    ).add_child(folium.Popup(site['Launch Site']))
    
    # Add a marker with the launch site name
    marker = folium.map.Marker(
        site_coordinate,
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % site['Launch Site'],
        )
    )
    
    site_map.add_child(circle)
    site_map.add_child(marker)

site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities

# Create a folium.PolyLine object using the coastline coordinates and launch site coordinate
# Define the coordinates as a list of [lat, lon] pairs

from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# First, let's pick a launch site (e.g., the first one in launch_sites_df)
launch_site_lat = launch_sites_df.iloc[0]['Lat']
launch_site_lon = launch_sites_df.iloc[0]['Long']
launch_site_name = launch_sites_df.iloc[0]['Launch Site']

print(f"Launch Site: {launch_site_name}")
print(f"Coordinates: ({launch_site_lat}, {launch_site_lon})")

# Find coordinates for coastline (you need to use MousePosition to get these)
# For CCAFS SLC-40, approximate coastline coordinates:
coastline_lat = 28.56367
coastline_lon = -80.56789

# Calculate distance to coastline
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
print(f"Distance to coastline: {distance_coastline:.2f} km")

# Create and add a folium.Marker on the selected closest coastline point
distance_marker = folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
    )
)
site_map.add_child(distance_marker)

# Draw a PolyLine between the launch site and coastline point
coastline_line = folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]], 
    weight=2, 
    color='blue'
)
site_map.add_child(coastline_line)

# Now add proximities for other features
# For railway (approximate coordinates near CCAFS SLC-40)
railway_lat = 28.57200
railway_lon = -80.58500
distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)
print(f"Distance to railway: {distance_railway:.2f} km")

# Add railway marker and line
railway_marker = folium.Marker(
    [railway_lat, railway_lon],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_railway),
    )
)
site_map.add_child(railway_marker)

railway_line = folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [railway_lat, railway_lon]], 
    weight=2, 
    color='red'
)
site_map.add_child(railway_line)

# For highway
highway_lat = 28.55500
highway_lon = -80.57500
distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)
print(f"Distance to highway: {distance_highway:.2f} km")

# Add highway marker and line
highway_marker = folium.Marker(
    [highway_lat, highway_lon],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_highway),
    )
)
site_map.add_child(highway_marker)

highway_line = folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [highway_lat, highway_lon]], 
    weight=2, 
    color='green'
)
site_map.add_child(highway_line)

# For city
city_lat = 28.54000
city_lon = -80.62000
distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)
print(f"Distance to city: {distance_city:.2f} km")

# Add city marker and line
city_marker = folium.Marker(
    [city_lat, city_lon],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_city),
    )
)
site_map.add_child(city_marker)

city_line = folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [city_lat, city_lon]], 
    weight=2, 
    color='purple'
)
site_map.add_child(city_line)

# Create a legend for the map
legend_html = '''
<div style="position: fixed; bottom: 50px; left: 50px; width: 200px; height: 130px; 
            border:2px solid grey; z-index:9999; font-size:14px;
            background-color:white; opacity:0.9;">
    &nbsp;<b>Legend</b><br>
    &nbsp;<span style="color:blue;">&nbsp;●&nbsp;</span> Coastline<br>
    &nbsp;<span style="color:red;">&nbsp;●&nbsp;</span> Railway<br>
    &nbsp;<span style="color:green;">&nbsp;●&nbsp;</span> Highway<br>
    &nbsp;<span style="color:purple;">&nbsp;●&nbsp;</span> City
</div>
'''
site_map.get_root().html.add_child(folium.Element(legend_html))

site_map
```
